```python
#importing necessary libraries

from pymysql import connect
import pandas as rsd
```

```python
#connect to the database

data_base= connect( host = 'localhost',
                  user= 'root', 
                  passwd= 'Neha081222#' )

#define cursor object

cur = data_base.cursor()

#define query to see all the databases available in the MYSQL server.

query= 'show databases'

#execute query

cur.execute(query)

#fetch all the databases
data_bases = cur.fetchall()

#iterate through all the databases
for data in data_bases:
    print(data)
```

```python
#now we navigate to the database we want

data_base = connect (host= 'localhost',
                    user= 'root',
                    passwd = 'Neha081222#', 
                    database= 'ipc_crimes2')

cur= data_base.cursor()
#check the available data

query = 'show tables'
cur.execute(query)

#fetch all the tables
tables= cur.fetchall()

#iterate through each table
for table in tables:
    print(table)
```

```python
query = 'SELECT `STATE/UT`, `YEAR`, DISTRICT, MURDER FROM (SELECT `STATE/UT`, `YEAR`, DISTRICT, MURDER, RANK() OVER (PARTITION BY `STATE/UT`, `YEAR` ORDER BY MURDER DESC) AS dis FROM ipc_crimes) AS district_data WHERE dis = 1'
cur.execute(query)
```

```python
#let's change our table to pandas dataframe

df= rsd.read_sql(query, data_base)
```
