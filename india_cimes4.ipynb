```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('C:/Users/mneha/OneDrive/Documents/GitHub/DSData/ipc_crimes.csv')
df
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab_enc= LabelEncoder()
```

```python
df2 = lab_enc.fit_transform(df['STATE/UT'])
pd.Series(df2)
```

```python
df['STATE/UT']= df2
df
```

```python
df3 = lab_enc.fit_transform(df['DISTRICT'])
pd.Series(df3)
```

```python
df['DISTRICT']= df3
df
```

```python
X= df[["DISTRICT", "TOTAL IPC CRIMES"]]
X.head()
```

```python
X.shape
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, random_state= 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.show()
    
    
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans= kmeans.fit_predict(X)
print(y_kmeans)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X,y_kmeans)
```

```python
cluster_1_District = df[y_kmeans ==0]
print(cluster_1_District)
```

```python
cluster_2_District = df[y_kmeans ==1]
print(cluster_2_District)
```

```python
cluster_3_District = df[y_kmeans ==2]
print(cluster_3_District)
```
